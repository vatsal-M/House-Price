# Data Cleaning

**I am going to drop columns with null value over half of the total data instances.**

```python
a=df_train.isnull().sum()
b=a.where(a>700).dropna()
b
```

```python
df_train=df_train.drop(['Alley','PoolQC','Fence','MiscFeature'],axis=1)
```

## Encoding the categorical features

```python
df_train=pd.get_dummies(df_train,drop_first=True)
df_train    
```

**One Categorical feature was left and hence I made an explicit function for encoding it**

```python
def my_ohe(str):
    "for getting encoded feaure data instances"
    dummies=pd.get_dummies(df_train[str],drop_first=True)
    merged=pd.concat([df_train,dummies],axis=1)
    final=merged.drop([str],axis=1)
    return final
```

```python
df_encoded=my_ohe('MSSubClass')
df_encoded
```

# Now I will use StandardScaler to Standardise the non Categorical Data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
x=df_encoded.drop(['SalePrice'],axis=1)
y=df_encoded['SalePrice']
scaled_x=StandardScaler().fit_transform(x)
X_train,X_test,y_train,y_test=train_test_split(scaled_x,y)
```

# Creating a function for GridSearch

```python
def GridSearch(cls,param,X_train=X_train,y_train=y_train):
    from sklearn.model_selection import GridSearchCV
    grid=GridSearchCV(cls,param_grid=param,cv=5)
    grid.fit(X_train,y_train)
    return grid
```

# KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
param={'n_neighbors':[1,2,4,5,6,8,10,15,20,40,60,100]}
knn_grid=GridSearch(knn,param)
knn_grid.score(X_test)
```
